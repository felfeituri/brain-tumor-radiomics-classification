```python
# List Group-Level Patient Folders
for folder in sorted(rootDir.iterdir()):
    if folder.is_dir():
        print("üìÅ", folder.name)
```

```python
# Count All Patient Folders
all_patient_dirs = [p for group in rootDir.iterdir() if group.is_dir()
                    for p in group.iterdir() if p.is_dir() and not p.name.startswith(".")]

print(f"üì¶ Total patient folders found: {len(all_patient_dirs)}")
```

```python
# Find all T1 images (supporting both .nii and .nii.gz)
t1_all = list(rootDir.rglob("*_t1_LPS_rSRI.nii")) + list(rootDir.rglob("*_t1_LPS_rSRI.nii.gz"))

# Find all masks (supporting both .nii and .nii.gz)
mask_all = list(rootDir.rglob("*_GlistrBoost_out-labels.nii")) + list(rootDir.rglob("*_GlistrBoost_out-labels.nii.gz"))

print(f"Found {len(t1_all)} T1 images")
print(f"Found {len(mask_all)} mask files")
```

```python
# Initialize PyRadiomics extractor
extractor = featureextractor.RadiomicsFeatureExtractor()
```

```python
# Initialize lists to store extracted features and skipped patients
all_features = []
skipped_patients = []
```

```python
# Loop through patient group folders
for group_folder in rootDir.iterdir():
    if group_folder.is_dir():
        for patient_dir in group_folder.iterdir():
            if not patient_dir.is_dir() or patient_dir.name.startswith("."):
                continue

            # Support both .nii and .nii.gz extensions
            t1_files = list(patient_dir.glob("*_t1_LPS_rSRI.nii")) + list(patient_dir.glob("*_t1_LPS_rSRI.nii.gz"))
            mask_files = list(patient_dir.glob("*_GlistrBoost_out-labels.nii")) + list(patient_dir.glob("*_GlistrBoost_out-labels.nii.gz"))

            if t1_files and mask_files:
                try:
                    t1_img = t1_files[0]
                    mask = mask_files[0]
                    result = extractor.execute(str(t1_img), str(mask))
                    result['PatientID'] = patient_dir.name
                    all_features.append(result)
                    print(f"‚úÖ Processed {patient_dir.name}")
                except Exception as e:
                    print(f"‚ö†Ô∏è Error processing {patient_dir.name}: {e}")
                    skipped_patients.append((patient_dir.name, str(e)))
            else:
                print(f"‚ùå Missing T1 or mask for {patient_dir.name}")
                skipped_patients.append((patient_dir.name, "T1 or mask not found"))
```

```python
# Convert to DataFrame
df = pd.DataFrame(all_features)
```
